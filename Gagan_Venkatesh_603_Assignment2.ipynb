#### Extracting the content to file1 and file2

```python
import pdfplumber

# path of the pdf file
pdf_path = "/content/Harry_Potter_(www.ztcprep.com).pdf"

# fnction to extract and clean text using pdfplumber
def extract_text(pdf_path, start_page, end_page):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page_num in range(start_page - 1, end_page):  # since pages are 0-indexed
            if page_num < len(pdf.pages):  # to avoid IndexError
                text += pdf.pages[page_num].extract_text() + "\n\n"

    # removing unnecessary line breaks and fix broken words
    text = text.replace("-\n", "").replace("\n", " ")

    return text

# extracting text for file1.txt (Pages 5949 to 5964)
file1_text = extract_text(pdf_path, 5949, 5964)

# extracting text for file2.txt (Pages 6101 to 6117)
file2_text = extract_text(pdf_path, 6101, 6117)

# saving extracted text to files
with open("file1.txt", "w", encoding="utf-8") as f1:
    f1.write(file1_text)

with open("file2.txt", "w", encoding="utf-8") as f2:
    f2.write(file2_text)

print("Text extraction complete.")
```
